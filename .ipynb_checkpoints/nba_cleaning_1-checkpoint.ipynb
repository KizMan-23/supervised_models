```python
#cross-checking for similar columns
def check_the_columns(players, mvps):
    PC = players.columns
    MC = mvps.columns
    for c in PC:
        if c in [d for d in MC]: 
            print (c)
        else:
            pass



def find_common_columns(df1, df2):
    common_columns = set(df1.columns) & set(df2.columns)
    return common_columns
```

```python
check_the_columns(players, mvp)
```

```python
del players["Rk"]
```

```python
players['Player'] = players['Player'].str.replace('*', '', regex=False)
```

```python
players.groupby(['Player', 'Year'])
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row["Tm"] = df.iloc[-1:]["Tm"]
        return row


players = players.groupby(['Player', 'Year']).apply(single_row)
```

```python
players.head(50)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
#some players are without teams. seems wrong but we will work with it 
players.apply(pd.isna).sum()
```

```python
all_stats = players.merge(mvp, how="outer", on = ["Player", "Year"])
```

```python
all_stats[["Pts Won", "Pts Max", "Share"]] = all_stats[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
#The empty teams will lead to inconsistency during the machine learning process
#Drop the rows with Teams empty
all_stats.apply(pd.isna).sum()
all_stats = all_stats.drop(all_stats[pd.isna(all_stats["Tm"])].index)
```

```python
#Reshaping the index
all_stats.index = range(all_stats.shape[0])
all_stats.isna().sum()


all_stats["FG%", "3P%", "FT%"] = all_stats["FG%", "3P%", "FT%"].fillna(0)  #columns are still objects
```
