```python
import pandas as pd
```

```python
df = pd.read_csv("premier_league_matches.csv")

df.sort_values("date_x")
df
```

```python
df.drop(df[['attendance', 'notes']], axis=1, inplace=True)

df = df.rename(columns={"date_x": "date", "year": "season"})

df['date'] = pd.to_datetime(df['date'])

df = df.sort_values('date')
df.reset_index(drop=True)


def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df.iloc[-1:]
        return row


df = df.groupby(['date', 'team'], group_keys=False).apply(single_row)

df.reset_index(drop=True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df[df.duplicated()]
```

```python
df['season'].value_counts()
```

```python
#To extract other date information

df['year'] = df['date'].dt.year
df['day_code'] = df['date'].dt.dayofweek
```

```python
#change some other important features

df['venue_code'] = df['venue'].astype('category').cat.codes
df['oppo_code'] = df['opponent'].astype('category').cat.codes
df['hour'] = df['time'].astype(str).str.slice(0,2).astype(int)
```

```python
df
```

```python
df = df.sort_values('date')
df = df.reset_index(drop=True)
```

```python
df
```

```python
#the target should be to predict win/loss of the next match. 

def add_target(df):
    df['target'] = (df['result'] == 'W').astype(int).shift(-1)
    return df

df = df.groupby(['team'], group_keys=False).apply(add_target)
df
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.shape
```

MODELLING 

```python
train  = df[df['date'] <= '2023-08-31']
test = df[df['date'] >= '2023-09-01']
```

```python
predictors = [
    'sot', 'gf', 'ga', 'dist','sh',
    'xg','xga', 'poss', 'pk', 'fk', 'hour',  
    'pkatt', 'day_code', 'oppo_code', 'venue_code'
]
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
```

```python
#Test different estimators to obtain the best.

model_params = {
    "knn" : {
        "model" : KNeighborsClassifier(n_jobs=2),
        "params" : {
            "n_neighbors" : [7,10, 15],
            "leaf_size" : [15, 20, 30],
            "weights" : ['uniform', 'distance']
        }
    },

    "hst" : {
        "model" : HistGradientBoostingClassifier(random_state=42),
        "params" : {
            "max_iter" : [75, 100, 150, 200],
            "max_depth" : [5, 10,15,20],
            "max_features" : [0.2, 0.5, 0.7, 1.0],
            "min_samples_leaf": [10,15,20,30]
        }
    },

    "rfc" :{
        "model" : RandomForestClassifier(random_state=42, n_jobs=2),
        "params" : {
            "n_estimators" : [50, 100, 200, 300],
            "max_depth": [5,10,15,20],
            "max_leaf_nodes": [2,5,7],
            "min_samples_leaf": [2,4,7,10]
        }
    }
}
```

```python
tscv = TimeSeriesSplit(n_splits=3)

scores = []

for model_name, mp in model_params.items():
    clf = RandomizedSearchCV(mp['model'], mp['params'], cv=tscv, random_state=42, n_iter=10)
    clf.fit(train[predictors], train['target'])
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
#to see the items within the dict
for model_name, mp in model_params.items():
    print(f"models: {model_name}")
    print(f" mps: {mp}")
```

```python
scores
#Histogram-based model gives the best results
```

```python
from sklearn.inspection import permutation_importance
```

```python

def find_permutation(model, predictors, target, column_name):
    train_impt = permutation_importance(
    model, predictors, target, random_state = 0, n_repeats = 42, n_jobs = 2
    )
    
    train_num = train_impt['importances_mean'].reshape(-1,1)
    train_num = pd.DataFrame(train_num, columns = column_name)

    features = predictors.columns
    features = pd.DataFrame(features, columns = ['features'])

    train_features = pd.concat([features, train_num], axis =1)
    top_train_features = train_features.sort_values(column_name, ascending=False).head(15)
    #least_train_features = train_features.sort_values(column_name, ascending=False).tail(20)

    return top_train_features
```

```python
top_train_features = find_permutation(hst, train[predictors], train['target'], ['train'])
top_test_features = find_permutation(hst, test[predictors], test['target'], ['test'])
```

```python
#the goal of the model is to predict the win/loss of the next match(target)
# we will imploy the current data which contains current win/loss to compute for upcoming matches

from sklearn.preprocessing import LabelEncoder
import numpy as np

df['result'] = LabelEncoder().fit_transform(df['result'])
```

```python
df['result']
```

```python
df_rolling = df[['gf', "ga", "result", 'sot', 'dist', 'sh', 'pk', 'pkatt', 'xg', 'xga', 'team', 'date']]
```

```python
df_rolling.sort_values('date')

df_rolling = df_rolling.drop('date', axis=1) #rolling method cannot be applied to datetime datatype
```

```python
df_rolling['team'] = LabelEncoder().fit_transform(df_rolling['team'])
```

```python
df_rolling
```

```python
def rolling_averages(df_rolling):
    rolling = df_rolling.rolling(3, closed='left').mean()
    return rolling
```

```python
roll = df_rolling.groupby(['team'], group_keys=False).apply(lambda x: rolling_averages(x))
```

```python
roll
```

```python
roll_cols = [f"{c}_rolling" for c in roll.columns]
```

```python
roll_cols
```

```python
roll.columns = roll_cols
```

```python
df = pd.concat([df, roll], axis=1)
```

```python
predictors += ['gf_rolling', 'ga_rolling', 'result_rolling',
 'sot_rolling', 'dist_rolling', 'sh_rolling', 'pk_rolling',
 'pkatt_rolling', 'xg_rolling','xga_rolling',]
```

```python
len(predictors)
```

```python
def Predictions(df, predictors):
    train  = df[df['date'] <= '2023-08-31']
    test = df[df['date'] >= '2023-09-01']
    hst.fit(train[predictors], train['target'])
    predss = hst.predict(test[predictors])
    predss = pd.DataFrame(predss, columns=['prediction'], index=test.index)
    results = pd.concat([test[['date','team', 'opponent', 'target']], predss], axis=1)

    precision = precision_score(results['target'], results['prediction'], zero_division=0)
    return results, precision
```

```python
result, precision = Predictions(df, predictors)
```

```python
#feed the model with pre-available information

def shift_column(df_1, col_name):
    next_col = df_1[col_name].shift(-1)


def add_column(df_1, col_name):
    return df_1.groupby("team", group_keys=False).apply(lambda x: shift_column(x, col_name))


df['opponent_next'] = add_column(df, 'opponent')
df['venue_next'] = add_column(df, "venue")
df['date_next'] = add_column(df, "date")
```

```python
df['oppo_next'] = df.groupby('team', group_keys=False).apply( lambda x: x['opponent'].shift(-1))

#the data set is not enough to shift necessary columns, delete Nans and fit the remaining to the model
```

```python
df['venue_next'] = df.groupby('team', group_keys=False).apply( lambda x: x['venue'].shift(-1))
df['date_next'] = df.groupby('team', group_keys=False).apply( lambda x: x['date'].shift(-1))
```
