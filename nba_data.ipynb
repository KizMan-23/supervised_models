DATA CLEANING


```python
df = pd.read_csv("C:/Users/Hp Pc/Downloads/Player Totals-checkpoint.csv")
```

```python
#mvp dataset
df_1 = pd.read_csv('C:/Users/Hp Pc/Downloads/Player Award Shares-checkpoint.csv')
df_1

df_1 = df_1[df_1['award'] == 'nba mvp']

df_1 = df_1[["season", "player", "pts_won", "pts_max", "share"]]
df_1.index = df_1['season']

df_1 = df_1.loc["2023":"1990"]

df_1.drop('season', axis = 1, inplace = True)
df_1
```

```python
df.dtypes
```

```python
del df['birth_year']
```

```python
df['age'] = df['age'].fillna(df['age'].mean)
df.fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
df['season'].unique()
```

```python
df['lg'].unique()
```

```python
#only NBA dataset is what I will need for the modelling
nba_players = df[df['lg'] == "NBA" ]
nba_players
print(nba_players.lg.unique())

#nba_players['season'] = pd.to_datetime(nba_players['season'])

nba_players.index = nba_players['season']
nba_players.index

nba_players.loc["2022":"1990"]

nba_players = nba_players.loc['2023':'1990']
nba_players

del nba_players['seas_id']
del nba_players['player_id']
del nba_players['season']
nba_players
```

```python
all_players = nba_players.merge(df_1, how = "outer", on = ['season', 'player'])

all_players

all_players[['pts_won', "pts_max", "share"]] = all_players[['pts_won', "pts_max", "share"]].fillna(0)

all_players[['player', 'tm']].head(50)

all_players.groupby(['season', 'player']).get_group((2022, 'Aaron Holiday'))

#single row function to deprecate players on multiple teams within a season into a single team
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row =  df[df['tm'] == 'TOT']
        row['tm'] = df.iloc[-1,:]['tm']
        return row

all_players = all_players.groupby(['season', 'player']).apply(single_row)

all_players.index = all_players.index.droplevel()

all_players.index = all_players.index.droplevel()

all_players[all_players['player'] == 'Aaron Holiday']
```

```python
#first-three nba players with the most points per season
pd.options.display.max_rows = 100
all_players.groupby(['season'], group_keys =False).apply(lambda x: x.sort_values('pts', ascending = False).head(3)).head(100)
```

```python
#Top three mvp players voted per season

all_players.groupby(['season'], group_keys=False).apply(
    lambda x: x.sort_values('share', ascending = False).head(3)).head(100)
```

DATA CLEANING - TEAMS


```python
team = pd.read_csv('C:/Users/Hp Pc/Downloads/Team Summaries-checkpoint.csv')

team = team[team['lg'] == 'NBA']

team.columns

team = team[['season', 'lg', 'team', 'abbreviation', 'w', 'l',
       'pw', 'pl', 'mov', 'sos', 'srs', 'o_rtg', 'd_rtg', 'n_rtg', 'pace',
       'f_tr', 'x3p_ar', 'ts_percent', 'e_fg_percent', 'tov_percent',
       'orb_percent', 'ft_fga']]

#select the season range as needed for the prediction
team = team[(team['season'] >= 1990) & (team['season'] <= 2023)]

team.isnull().sum()

#find out the teams without abbreviation
team[(team['abbreviation'].isnull())][['team', 'abbreviation']]
```

```python
team[team['team'] == 'League Average']

#the team without abbrev is not an actual team. so drop the team
team = team.drop(team[team['team'] == 'League Average'].index)

team.index = range(team.shape[0])
team.isnull().sum()

print(team['team'].unique())
print(team['team'].value_counts())
print(team['season'].unique())
print(team['team'].nunique())
```

```python
team.rename(columns ={'abbreviation': 'tm'}, inplace = True)
team
```
